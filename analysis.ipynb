```python
import math
import math_pi

# Step 1: Create mapping from '00' to '99' using ASCII 33132
double_digit_map = {str(i).zfill(2): chr(33 + i) for i in range(100)}

# Step 2: Get decimal part of pi
pi_str = math_pi.pi(a=1, b=1000000)
decimal_digits = pi_str.split('.')[1]

# Step 3: Process every 2 digits
pairs = [decimal_digits[i:i+2] for i in range(0, len(decimal_digits) - 1, 2)]

# Step 4: Map to ASCII characters
mapped_chars = ''.join(double_digit_map[p] for p in pairs if p in double_digit_map)

# Step 5: Break into chunks of floor(exp(pi)) = 23
chunk_length = int(math.floor(math.exp(math.pi)))  #  23
chunks = [mapped_chars[i:i + chunk_length] for i in range(0, len(mapped_chars), chunk_length)]

# Step 6: Print each chunk (stacked rows)
for chunk in chunks:
    print(chunk)
```
